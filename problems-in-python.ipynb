## **To check in a DataFrame in Python whether the entries of two columns are equal, even if the headings of these columns are not equal, you can perform the following steps**

```python
import pandas as pd

# Example DataFrame with possibly different headings

data = {
    'column_a':[1,2,3,4,5],
    'column_b':[1,4,3,8,5],
    'column_c':[10,20,30,40,50]
}

df = pd.DataFrame(data)

# Find the columns whose values are to be compared (here: 'column_a' and 'column_b')

column_a = df.columns[df.columns.str.contains('column_a', case = False)].values[0]
column_b = df.columns[df.columns.str.contains('column_b', case = False)].values[0]

# Check if the values in the two columns are the same

same_columns = (df[column_a] == df[column_b])

# Output the result
print(same_columns)
```
